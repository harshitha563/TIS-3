```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create 30 days sales data
sales = pd.DataFrame({
    "Day": range(1, 31),
    "Sales": [12000,15000,13000,16000,17000,18000,14000,15500,16500,17500,
              19000,20000,21000,22000,23000,24000,25000,26000,27000,28000,
              29000,30000,31000,32000,33000,34000,35000,36000,37000,38000]
})

# Mean, Median, Mode
print("Mean:", sales["Sales"].mean())
print("Median:", sales["Sales"].median())
print("Mode:", sales["Sales"].mode()[0])

# Standard Deviation & Variance
print("Standard Deviation:", sales["Sales"].std())
print("Variance:", sales["Sales"].var())

# Correlation (Day vs Sales)
correlation = sales["Day"].corr(sales["Sales"])
print("Correlation between Day and Sales:", correlation)
```
