```python
### 1.create a dataframe where it consists of columns like name,dept,salary filter by department group by department
```

```python
import pandas as pd

# Create employee dataframe
employees = pd.DataFrame({
    "name": ["Asha", "Rahul", "Kiran", "Sneha", "Arjun", "Meera"],
    "dept": ["HR", "IT", "IT", "Finance", "HR", "Finance"],
    "salary": [40000, 60000, 65000, 50000, 42000, 52000]
})

print("Employee DataFrame:")
print(employees)
```

```python
it_employees = employees[employees["dept"] == "IT"]

print("\nFiltered Employees (IT Department):")
print(it_employees)
```

```python
grouped_data = employees.groupby("dept")["salary"].mean().reset_index()

print("\nAverage Salary by Department:")
print(grouped_data)
```

```python
### 2.create a dataframe where it consistss of columns like product id,product name,quantity and sales find the top 5 product based on the total sales
```

```python
import pandas as pd

# Create Product DataFrame
products = pd.DataFrame({
    "product_id": [101, 102, 103, 104, 105, 106, 107],
    "product_name": ["Laptop", "Mouse", "Keyboard", "Monitor", "Printer", "Tablet", "Mobile"],
    "quantity": [10, 50, 40, 20, 15, 25, 30],
    "sales": [500000, 25000, 40000, 150000, 75000, 200000, 350000]
})

print("Product DataFrame:")
print(products)
```

```python
top_5_products = products.sort_values(by="sales", ascending=False).head(5)

print("\nTop 5 Products Based on Sales:")
print(top_5_products)
```

##### What is Pandas? What are the important features of Pandas?
Pandas is an open-source Python library used for data analysis and data manipulation. It provides two main data structures: Series (1D) and DataFrame (2D). Important features include handling missing data, data alignment and indexing, grouping and aggregation, filtering and sorting, merging and joining datasets, time series support, and reading/writing data from formats like CSV, Excel, SQL, and JSON.

##### How do you handle large datasets efficiently in Pandas?
To handle large datasets efficiently, use chunksize while reading files, load only required columns using usecols, convert columns to proper data types like category, use vectorized operations instead of loops, filter data using query(), and consider tools like Dask or PySpark for very large datasets.

##### What are the different methods available in Series and their purpose?
Common Series methods include head() to view first rows, tail() to view last rows, mean() to calculate average, sum() for total, max() and min() for highest and lowest values, value_counts() to count unique values, unique() to get distinct values, isnull() to check missing values, fillna() to replace missing values, and astype() to change data type.

##### How do you optimize memory usage in Pandas?
Memory can be optimized by converting object columns to category type, downcasting numeric columns using pd.to_numeric(), removing unnecessary columns, checking memory usage with memory_usage(), and reading large files in chunks.

##### Explain about groupby method in Pandas.
The groupby() method is used to split data into groups based on a column and apply aggregation functions like sum(), mean(), count(), max(), or min(). It is mainly used for summarizing and analyzing grouped data.

##### What is the difference between Series and DataFrame?
A Series is a one-dimensional data structure that represents a single column of data. A DataFrame is a two-dimensional table with multiple columns and rows. Series contains only one data type, whereas a DataFrame can contain multiple data types.
