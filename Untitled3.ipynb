```python
import numpy as np
import matplotlib.pyplot as plt

mean = 0
std = 1
size = 100

prices = np.cumsum(np.random.normal(mean, std, size)) + 100

plt.plot(prices)
plt.title("Stock Price Movement")
plt.xlabel("Days")
plt.ylabel("Stock Price")
plt.grid(True)
plt.show()
```

```python
# 7. Sales Forecast vs Actual --- np.linspace(), np.random.normal()

days = np.arange(30)
forecast = np.linspace(100, 200, 30)
actual = forecast + np.random.normal(0, 10, 30)

plt.plot(days, forecast, label="Forecast")
plt.plot(days, actual, label="Actual")
plt.title("Sales Forecast vs Actual")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)
plt.show()
```

1. What is Matplotlib? Explain about it.
Matplotlib is a Python data visualization library used to create charts and graphs. It helps in representing data visually so that trends, patterns, and relationships can be easily understood. It supports various types of plots such as line plots, bar charts, histograms, scatter plots, and pie charts. It is widely used in data science, machine learning, and research projects for data analysis and presentation.

2. Difference between plot() and scatterplot()
The plot() function is mainly used to create line graphs where data points are connected with lines, making it suitable for showing trends over time. The scatter() function is used to create scatter plots where individual data points are displayed as dots and are not connected. Line plots are best for trend analysis, while scatter plots are best for showing relationships between two variables.

3. What is pyplot? Explain about it.
Pyplot is a module in Matplotlib that provides a simple interface for creating plots and charts. It is usually imported as import matplotlib.pyplot as plt. It contains functions like plot(), scatter(), hist(), bar(), and show() which help in creating and displaying graphs easily. Pyplot makes plotting similar to MATLAB style and is beginner-friendly.

4. How to add labels and title?
Labels and titles can be added using the following functions:
plt.title("Title Name")  Adds title to the graph
plt.xlabel("X-axis label")  Adds label to x-axis
plt.ylabel("Y-axis label")  Adds label to y-axis
Adding labels and titles makes the graph clear and easy to understand.

5. What is the difference between figure and axes?
A Figure is the entire canvas or window that holds the complete plot. An Axes is the actual area inside the figure where the data is plotted. A figure can contain multiple axes. In simple words, the figure is the outer container, and axes is the graph area inside it.

6. How to create multiple subplots?
Multiple subplots can be created using:
plt.subplot(rows, columns, position)
plt.subplots()
These functions allow us to create more than one graph in the same figure. It is useful when comparing multiple datasets side by side.

7. What is the use of marker? What are the different markers we can use?
Markers are used to highlight individual data points in a plot. They make the graph clearer and easier to read.
Some common markers are:
'o'  Circle
's'  Square
'^'  Triangle
'*'  Star
'+'  Plus
'x'  Cross

8. How to customize colors?
Colors can be customized using the color parameter in plotting functions. Example:
plt.plot(x, y, color='red')
We can use:
Color names (red, blue, green)
Hex codes (#FF5733)
RGB values
Customizing colors improves visualization and presentation.

9. How to handle large amount of data while visualization?
To handle large datasets:
Reduce marker size
Use transparency (alpha parameter)

Use sampling (plot only part of data)

Use efficient plot types like line plots instead of scatter when needed

Use interactive visualization tools

These techniques improve performance and clarity.

10. Which plot is used for performing trends in the data?

A Line Plot is used to show trends in data. It is especially useful for time-series data such as stock prices, temperature changes, and sales growth over time.

11. Which plot is used for data distributions?

A Histogram is used to show data distribution. It displays how data is spread across different ranges and helps in identifying patterns like skewness and spread.

12. How to display grid?

A grid can be displayed using:
plt.grid(True)

The grid improves readability and helps in comparing values easily.

13. How to change line color?

Line color can be changed using:
plt.plot(x, y, color='green')

We can choose any color name or hex color code.

14. What are the different styles available in line?

Different line styles available in Matplotlib are:

'-'  Solid line

'--'  Dashed line

':'  Dotted line

'-.'  Dash-dot line

Line styles help differentiate multiple lines in the same graph.
